<h4 align = 'center'><b>Data Cleaning</b></h4>

```python
print(data.isnull().sum())
```

```python
data.dropna(axis=1,how='all',inplace=True)
```

```python
data.dropna(thresh=5,inplace=True)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.columns = data.columns.str.strip().str.lower().str.replace(" ", "_").str.replace(r"[^a-z0-9_]", "", regex=True)
```

<h5 align = 'center'><b> Encode Categorical Data</b><h5>

```python
education_map = {
    "Doctoral degree": 4,
    "Master‚Äôs degree": 3,
    "Bachelor‚Äôs degree": 2,
    "Some college/university study without earning a degree": 1,
    "I prefer not to answer": 0,
    "None": 0
}

data['education_encoded'] = data['q4'].map(education_map)
```

```python
data
```

 <h5 align = 'center'><b>Insight Generation</b></h5>

##### **Top 5 Countries by Respondents**

```python
data['q3'].value_counts().head(5)
```

##### **Most Used Programming Languages**

```python
language_cols = [col for col in data.columns if 'q7_part' in col]
data_lang = data[language_cols]
data_lang = data_lang.apply(pd.Series.value_counts).sum(axis=1).sort_values(ascending=False)
data_lang.head(5)
```

##### **Salary Range by Education Level**

```python
data[['q4', 'q25']].groupby('q4')['q25'].value_counts().unstack().fillna(0)
```

<h5 align = 'center'><b>Visualization</b></h5>

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

<h5><b>Top 5 Programming Languages</b></h5>

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for top 5 languages
plt.figure(figsize=(10, 6))
sns.barplot(x=data_lang.head(5).values, y=data_lang.head(5).index, palette='viridis')
plt.title("Top 5 Programming Languages (2017‚Äì2021)")
plt.xlabel("Number of Users")
plt.ylabel("Programming Language")
plt.tight_layout()
plt.show()
```

<h5><b>Education Level Distribution</b></h5>

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='education_encoded', data=data, palette='magma', order=sorted(data['education_encoded'].dropna().unique(), reverse=True))
plt.title("Education Level Distribution")
plt.xlabel("Count")
plt.ylabel("Education Level Code (0=Low, 4=High)")
plt.tight_layout()
plt.show()
```

<h5><b>Programming Language Usage by Education Level</b></h5>

```python
# Melt language columns to long format for grouping
data_lang_melted = data[language_cols + ['education_encoded']].melt(id_vars='education_encoded', value_name='Language').dropna()

plt.figure(figsize=(12, 6))
sns.countplot(data=data_lang_melted, x='Language', hue='education_encoded', order=data_lang_melted['Language'].value_counts().index[:5])
plt.title("Top 5 Language Usage by Education Level")
plt.xlabel("Programming Language")
plt.ylabel("Count")
plt.legend(title="Education Level")
plt.tight_layout()
plt.show()
```

#### ‚úÖ Insights and Key Findings

- Python consistently dominated as the top programming language, followed by SQL, R, and C/C++, confirming its relevance in data science and ML roles.
- Most respondents held a Master‚Äôs or Bachelor‚Äôs degree, showing the importance of formal education in this field.
- Users with higher education levels showed more diverse language usage, likely due to broader exposure and professional demands.


#### ‚úÖ Conclusion
This analysis reflects the evolving trends in the data science community from 2017 to 2021. It highlights that:

- Python remains the most crucial language, valued by professionals of all educational backgrounds.
- Education level influences the depth and diversity of tool usage.
- The Kaggle community is primarily composed of well-educated professionals, reinforcing the competitive nature of the data science field.
